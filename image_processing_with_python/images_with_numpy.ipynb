### Images of different formats red green and blue

```python
import matplotlib.pyplot as plt 
from skimage import io
benedict = plt.imread('original.jpg')
type(benedict)
red = benedict[:,:,0]
plt.imshow(red,cmap='gray')
plt.title('Red')
plt.show()
```

```python
import matplotlib.pyplot as plt 
from skimage import io
benedict = plt.imread('original.jpg')
type(benedict)
blue = benedict[:,:,2]
plt.imshow(blue,cmap='gray')
plt.title('Blue')
plt.show()
```
